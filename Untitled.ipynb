#Real time video demo for Face Recognition

```python
import mysql.connector
from mysql.connector import Error
import tkinter as tk
from tkinter import Label, Button, Entry, messagebox
from PIL import Image, ImageTk
import threading
import queue
import collections
import cv2
from deepface import DeepFace

# MySQL Database Connection
def create_connection():
    connection = None
    try:
        connection = mysql.connector.connect(
            host="localhost",
            user="root",  # Replace with your MySQL username
            password="Deepika123@",  # Replace with your MySQL password
            database="Emotion_DetectionDB"
        )
        print("Connection to MySQL DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")
    return connection

def fetch_query(connection, query, data):
    cursor = connection.cursor(dictionary=True)
    try:
        cursor.execute(query, data)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
        return None

# Queue for frames
frame_queue = queue.Queue(maxsize=5)

# Global flag to control processing loop
processing = False

# Dictionary to store emotions and their counts for smoothing
emotion_counts = collections.defaultdict(lambda: collections.deque(maxlen=10))

def process_frames():
    global processing
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    while processing:
        if not frame_queue.empty():
            frame = frame_queue.get()
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = faceCascade.detectMultiScale(gray, 1.1, 4)
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                face_crop = frame[y:y + h, x:x + w]
                try:
                    results = DeepFace.analyze(face_crop, actions=['emotion'], enforce_detection=False)
                    if isinstance(results, list):
                        result = results[0]
                    else:
                        result = results

                    dominant_emotion = result['dominant_emotion']
                    emotion_counts[(x, y, w, h)].append(dominant_emotion)
                    dominant_emotion = max(set(emotion_counts[(x, y, w, h)]), key=emotion_counts[(x, y, w, h)].count)

                    font = cv2.FONT_HERSHEY_SIMPLEX
                    cv2.putText(frame, dominant_emotion, (x, y - 10), font, 0.9, (0, 255, 0), 2, cv2.LINE_AA)
                except Exception as e:
                    print("Error:", e)
            cv2.imshow('Demo video', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                processing = False
                break

def read_frames(cap, frame_skip=10):
    global processing
    frame_count = 0

    while processing:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_skip == 0:
            if not frame_queue.full():
                frame_queue.put(frame)
        frame_count += 1
        cv2.waitKey(1)

    cap.release()
    cv2.destroyAllWindows()
    processing = False

def analyze_emotions_webcam():
    global processing
    processing = True
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        raise IOError("Cannot open webcam.")

    threading.Thread(target=process_frames, daemon=True).start()
    threading.Thread(target=read_frames, args=(cap,), daemon=True).start()

# Login Functionality
def login():
    username = entry_username.get()
    password = entry_password.get()
    email = entry_email.get()

    if not username or not password or not email:
        messagebox.showerror("Error", "All fields are required!")
        return

    connection = create_connection()
    query = "SELECT * FROM usersLogin WHERE username = %s AND password = %s AND email = %s"
    data = (username, password, email)
    user = fetch_query(connection, query, data)

    if user:
        messagebox.showinfo("Success", "Login Successful!")
        login_screen.destroy()
        emotion_detection_screen()
    else:
        messagebox.showerror("Error", "Invalid credentials")

def emotion_detection_screen():
    root = tk.Tk()
    root.title("Emotion Detection")
    root.geometry("1000x600")

    try:
        image = Image.open("C:/Users/kmdee/OneDrive/Pictures/Documents/Desktop/background/Anaconda_face_emotion_recognition/background12.png")
        image = image.resize((1000, 600), Image.Resampling.LANCZOS)
        background_image = ImageTk.PhotoImage(image)
        background_label = Label(root, image=background_image)
        background_label.image = background_image  # Keep a reference to the image
        background_label.place(relwidth=1, relheight=1)
        print("Background image loaded successfully for emotion detection screen.")
    except Exception as e:
        print(f"Error loading background image for emotion detection screen: {e}")

    button_style = {
        "font": ("Helvetica", 16),
        "bg": "#4CAF50",
        "fg": "white",
        "activebackground": "#45a049",
        "activeforeground": "white",
        "bd": 10,
        "relief": "raised"
    }

    btn_webcam = Button(root, text="Webcam Face Emotion Detection", command=analyze_emotions_webcam, **button_style)
    btn_webcam.place(relx=0.1, rely=0.5, anchor='w')

    root.mainloop()

# Tkinter setup for login
login_screen = tk.Tk()
login_screen.title("Admin Login")
login_screen.geometry("1000x600")

# Load the background image for the left side of the login screen
try:
    login_image = Image.open("C:/Users/kmdee/OneDrive/Pictures/Documents/Desktop/background/Anaconda_face_emotion_recognition/imagenew.jpg")
    login_image = login_image.resize((500, 600), Image.Resampling.LANCZOS)
    login_background = ImageTk.PhotoImage(login_image)
    login_background_label = Label(login_screen, image=login_background)
    login_background_label.image = login_background  # Keep a reference to the image
    login_background_label.place(relx=0, rely=0, relwidth=0.5, relheight=1)
    print("Background image loaded successfully for login screen.")
except Exception as e:
    print(f"Error loading login background image: {e}")

# Create a frame for the login form on the right side
login_frame = tk.Frame(login_screen, bg='white', bd=5)
login_frame.place(relx=0.75, rely=0.5, anchor='center', relwidth=0.4, relheight=0.8)

label_style = {
    "font": ("lucida calligraphy", 12),
    "bg": "white",
    "fg": "black"
}

entry_style = {
    "font": ("Helvetica", 12),
    "bg": "white",
    "fg": "black",
    "bd": 2,
    "relief": "solid",
    "insertbackground": "black"
}

Label(login_frame, text="Admin Login", font=("lucida calligraphy", 18, "bold"), bg="white", fg="black").pack(pady=20)

Label(login_frame, text="Username", **label_style).pack(pady=10)
entry_username = Entry(login_frame, **entry_style)
entry_username.pack(pady=10)

Label(login_frame, text="Password", **label_style).pack(pady=10)
entry_password = Entry(login_frame, show='*', **entry_style)
entry_password.pack(pady=10)

Label(login_frame, text="Email-id", **label_style).pack(pady=10)
entry_email = Entry(login_frame, **entry_style)
entry_email.pack(pady=10)

button_style = {
    "font": ("lucida calligraphy", 15),
    "bg": "#4CAF50",
    "fg": "white",
    "activebackground": "#45a049",
    "activeforeground": "white",
    "bd": 2,
    "relief": "raised"
}

Button(login_frame, text="Login", command=login, **button_style).pack(pady=30)

login_screen.mainloop()
```
